```python
import matplotlib.pyplot as plt

x = [2**n for n in range(5, 15)]
x.append(20000)

with open('mean_times5.txt', 'r') as f:
    y1 = list(map(float, f.read().replace(',', ' ').split()))
    
with open('mean_times10.txt', 'r') as f:
    y2 = list(map(float, f.read().replace(',', ' ').split()))

with open('mean_times15.txt', 'r') as f:
    y3 = list(map(float, f.read().replace(',', ' ').split()))
    
with open('mean_times20.txt', 'r') as f:
    y4 = list(map(float, f.read().replace(',', ' ').split()))
```

```python
plt.figure(figsize=(10, 8))
plt.plot(x, y1, label ='k=5')
plt.plot(x, y2, label ='k=10')
plt.plot(x, y3, label ='k=15')
plt.plot(x, y4, label ='k=20')

plt.xlabel("Количество вершин в графе")
plt.ylabel("Время (s)")
plt.legend()
plt.title('Зависимость среднего времени построения графа от количества соседей(k) и размера графа(n)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(14, 12))
plt.suptitle('Зависимость времени работы каждого из алгоритмов от размера графа')
x = [2**n for n in range(5, 12)]
for id, k in enumerate([5, 10, 15, 20]):
    y = [[] for n in range(4)]
    for method in [0, 1]:
        for threads in [0, 1]:
            with open(f'algo{method}{threads}{k}.txt', 'r') as f:
                y[method * 2 + threads]= list(map(float, f.read().replace(',', ' ').split()))
    plt.subplot(2, 2, id+1)
    plt.plot(x, y[0], label ='IncSingle')
    plt.plot(x, y[1], label ='IncMulti')
    plt.plot(x, y[2], label ='RieserSignle')
    plt.plot(x, y[3], label ='RieserMulti')
    plt.legend()
    plt.xlabel('Размер графа')
    plt.title(f'k={k}')
    plt.ylabel('Время (c)')
plt.tight_layout()

plt.show()
```
