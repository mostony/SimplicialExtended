## Graph visualization

```python
import networkx as nx
G = nx.Graph()
```

```python
from dataset_loader import DatasetLoader

sets, true_labels = DatasetLoader.load_dataset("contact-high-school")
```

```python
from network_builder import GraphBuilder
```

```python
graph_builder = GraphBuilder(sets, True)
graph = graph_builder.build()
```

```python
for edge in graph_builder.edges:
	v, u = edge
	G.add_edge(v, u)
```

```python
for i, label in enumerate(true_labels):
	G.nodes[i]['label'] = label
```

```python
import matplotlib.pyplot as plt
pos = nx.kamada_kawai_layout(G)
labels = {i: G.nodes[i]['label'] for i in G.nodes}
plt.figure(figsize=(12, 13))

colors = [
    'red', 'blue', 'green', 'purple', 'orange',
    'cyan', 'magenta', 'yellow', 'brown', 'pink', 'lime'
]
tmp = [labels[n] for n in G.nodes()]
node_colors = [colors[labels[n] - 1] for n in G.nodes()]
nx.draw(G, pos=pos, with_labels=True, node_color=node_colors)  

plt.show()
```
